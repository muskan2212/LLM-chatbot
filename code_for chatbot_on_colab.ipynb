```python
import PyPDF2

pdfFileObj = open('/content/muskanLOR.pdf', 'rb')
# creating a pdf reader object
pdfReader = PyPDF2.PdfReader(pdfFileObj)
text = ""
for page in pdfReader.pages[:4]:
  text += page.extract_text()

# closing the pdf file object
pdfFileObj.close()
```

```python
#get chunks of text
```

```python
from langchain.text_splitter import CharacterTextSplitter
```

```python
text_splitter = CharacterTextSplitter(separator="\n", chunk_size=1000, chunk_overlap=200, length_function=len)
chunks = text_splitter.split_text(text)
print(len(chunks))
```

```python
#get vectorstores
```

```python
from langchain.vectorstores import FAISS
```

```python
from langchain.embeddings import OpenAIEmbeddings,HuggingFaceInstructEmbeddings
```

```python
from langchain.embeddings import HuggingFaceInstructEmbeddings

embeddings = HuggingFaceInstructEmbeddings(model_name="hkunlp/instructor-xl")

vectorstore = FAISS.from_texts(texts=chunks, embedding=embeddings)
print(vectorstore)
```

```python
from langchain.memory import ConversationBufferMemory
from langchain.chains import ConversationalRetrievalChain
```

```python
from langchain.llms import HuggingFaceHub
```

```python
from google.colab import userdata
token=userdata.get('Token')
```

```python
TOKEN=token
```

```python
llm = HuggingFaceHub(repo_id="google/flan-t5-xxl",huggingfacehub_api_token=TOKEN, model_kwargs={"temperature":0.5, "max_length":512})
memory = ConversationBufferMemory(memory_key='chat_history', return_messages=True)

conversation_chain = ConversationalRetrievalChain.from_llm(llm=llm, retriever=vectorstore.as_retriever(), memory=memory)
```

```python
from langchain.chains.question_answering import load_qa_chain
```

```python
# llm=HuggingFaceHub(repo_id="google/flan-t5-xxl",huggingfacehub_api_token=TOKEN, model_kwargs={"temperature":1, "max_length":1000000})
chain = load_qa_chain(llm, chain_type="stuff")

#QUERYING
query = "what is role"
docs = vectorstore.similarity_search(query)
chain.run(input_documents=docs, question=query)
```
